```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/kushalrathod/Desktop/Projects/STT_lab9/TypeMetrics.csv')
df.head()
```

```python
lcom1_90th_percentile = df['LCOM1'].quantile(0.9)
print(f"90th percentile of LCOM1: {lcom1_90th_percentile}")

# Filter classes with LCOM1 above the 90th percentile
high_lcom1_classes = df[df['LCOM1'] > lcom1_90th_percentile]

# Display the classes with high LCOM1 values
print("Classes with high LCOM1 values:")
display(high_lcom1_classes[['Package Name','Type Name', 'LCOM1', "LCOM2","LCOM3","LCOM4","LCOM5","YALCOM"]].sort_values(by='LCOM1', ascending=False, ignore_index=True))
```
